## Logistics Regression

```python
# split the training and testing data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop("output", axis = 1), df.output, test_size = 0.3, random_state = 100)
```

```python
# validating if the data are correctly splitted
X_train, y_train, X_test, y_test
```

```python
# training the model with X_train and y_train
from sklearn.linear_model import LogisticRegression
logmodel = LogisticRegression()
logmodel.fit(X_train, y_train)
```

```python
predictions = logmodel.predict(X_test)
```

```python
predictions, len(predictions)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
confusion_matrix(y_test, predictions)
```

```python
print(classification_report(y_test, predictions))
```

```python
confusion_df = pd.DataFrame(confusion_matrix(y_test, predictions), columns=["Predicted class "+
                           str(class_name) for class_name in [0, 1]], index = ["Class " + 
                           str(class_name) for class_name in [0, 1]])
confusion_df
```

```python
# print the coefficient and intercept of the model
print(logmodel.coef_)
print(logmodel.intercept_)
```

```python
# fetching the precision, recall, fscore, support, accuracy_score in a more usable mode
# we can use these value to store in a variable for future reference

from sklearn.metrics import precision_recall_fscore_support, accuracy_score
print(precision_recall_fscore_support(y_test, predictions), '\n')
accuracy_score_logisticRegression = accuracy_score(y_test, predictions)
print(accuracy_score_logisticRegression)
```

## K Nearest Neighbour

```python
# split the training and testing data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop("output", axis = 1), df.output, test_size = 0.3, random_state = 101)
```

```python
X_train, X_test, y_train, y_test
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
classifier = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
```

```python
classifier.fit(X_train, y_train)
```

```python
predictions = classifier.predict(X_test)
```

```python
confusion_matrix(y_test, predictions)
```

```python
print(classification_report(y_test, predictions))
```

```python
confusion_df = pd.DataFrame(confusion_matrix(y_test, predictions), columns=["Predicted class "+
                           str(class_name) for class_name in [0, 1]], index = ["Class " + 
                           str(class_name) for class_name in [0, 1]])
confusion_df
```

```python
precision_recall_fscore_support(y_test, predictions)
```

```python
accuracy_score_knn = accuracy_score(y_test, predictions)
print(accuracy_score_knn)
```

```python
plt.title("Accuracy Measure")
sns.barplot(x = ['Logistic Regression', 'KNN'], y = [accuracy_score_logisticRegression, accuracy_score_knn])
plt.plot()
```

```python
print("Accuracy of Logistic Regression Model: ", accuracy_score_logisticRegression)
print("Accuracy of KNN Model: ", accuracy_score_knn)
```

### Decision Tree

```python
# split the training and testing data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop("output", axis = 1),
                                                    df.output, test_size = 0.3,
                                                    random_state = 80)
```

```python
from sklearn.tree import DecisionTreeClassifier
decision_tree_model = DecisionTreeClassifier(criterion = 'gini', random_state = 100,
                                             max_depth = 5, min_samples_leaf = 5)
```

```python
decision_tree_model.fit(X_train, y_train)
```

```python
from sklearn import tree
with open("decision_tree_model_graph.txt", "w") as f:
    f = tree.export_graphviz(decision_tree_model, out_file = f)
```

```python
# from sklearn.externals.six import StringIO 
from io import StringIO
from IPython.display import Image 
from sklearn.tree import export_graphviz
import pydotplus
dot_data = StringIO()
export_graphviz(decision_tree_model, out_file=dot_data,filled=True, rounded=True, special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue()) 
Image(graph.create_png())
```

```python
predictions = decision_tree_model.predict(X_test)
```

```python
confusion_matrix(y_test, predictions)
```

```python
print(classification_report(y_test, predictions))
```

```python
confusion_df = pd.DataFrame(data = confusion_matrix(y_test, predictions), 
                           columns = ["Predicted Class " + str(class_no) for class_no in [0, 1]],
                            index = ["Class " + str(class_no) for class_no in [0, 1]])
```

```python
confusion_df
```

```python
predictions
```

```python
precision_recall_fscore_support(y_test, predictions)
```

```python
accuracy_score_DecisionTree = accuracy_score(y_test, predictions)
```

```python
accuracy_score_DecisionTree
```

```python
plt.title("Accurace Measure")
sns.barplot(y = [accuracy_score_logisticRegression, accuracy_score_knn, accuracy_score_DecisionTree], 
           x = ["Logistics Regression", "KNN", "Decision Tree"])
print("Logistics Regression score:", round(accuracy_score_logisticRegression, 2))
print("KNN score:", round(accuracy_score_knn, 2))
print("Decision tree score:", round(accuracy_score_DecisionTree, 2))
plt.show()
```
